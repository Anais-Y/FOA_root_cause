```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
file = '../Ref_2351573/Avant_cluster2023-07-24_ALL.xlsx'
```

```python
data = pd.read_excel(file)
data_key = data[['AURORA      102', 'region_description']]
data_key.loc[data_key['region_description']!=3, 'region_description']=0
data_key.loc[data_key['region_description']==3, 'region_description']=1
```

```python
data_key.corr()
```

```python
data = pd.read_excel(file)
data_key = data[['PILOT', 'Local Courier', 'NIPPON', 'FEDEX', 'DHL', 'region_description']]
data_key['USCAN'] = data_key['region_description']==3
data_key['Intercontinental'] = data_key['region_description']==2
cor = data_key[['PILOT', 'Local Courier', 'NIPPON', 'FEDEX', 'DHL','Intercontinental', 'USCAN']].corr()
sns.heatmap(cor,annot=True, cmap="BuPu")
```
